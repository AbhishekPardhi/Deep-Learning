```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("House_prediction.csv")
grouped_data = df.groupby(["city"])[list(df.head(0))[2:]].mean()
grouped_data
```

```python
for i in range(grouped_data.shape[1]):
    plt.figure(figsize=(16, 9))
    plt.plot(grouped_data.index.tolist(),grouped_data.loc[:,list(grouped_data.head(0))[i]])
    plt.ylabel(list(grouped_data.head(0))[i])
    plt.show()
```

Here we can understand the correlation using heatmaps as follows:
Higher the value means variables are highly correlated.

```python
sns.set(rc = {'figure.figsize':(16,9)})
ax = sns.heatmap(grouped_data.corr(), annot=True)
```
